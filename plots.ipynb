```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df_pi = pd.read_csv("results/pi.txt")
```

```python
df_pi
```

```python
plt.plot(df_pi['threads'], df_pi['time_seconds'] * 1e3, marker='o', linestyle='-')

plt.xlabel("Num Threads", fontsize=14)
plt.ylabel("Time (ms)", fontsize=14)

plt.savefig("results/pi.png", bbox_inches='tight')
```

```python
df_car = pd.read_csv("results/car_save_frequency.txt", sep=" ")
df_car
plt.plot(df_car['save_every'], df_car['runtime_sec'], marker='o', linestyle='-')

plt.xlabel("Save Every Nth Frame", fontsize=14)
plt.ylabel("Time (sec)", fontsize=14)

plt.savefig("results/car_save_frequency.png", bbox_inches='tight')
```

```python
df_axisb = pd.read_csv("results/axisb.txt", sep=",")
df_axisb
plt.plot(df_axisb['threads'], df_axisb['time_seconds'], marker='o', linestyle='-')

plt.xlabel("Num Threads", fontsize=14)
plt.ylabel("Time (sec)", fontsize=14)

plt.savefig("results/axisb.png", bbox_inches='tight')
```
