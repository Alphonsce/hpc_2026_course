```python
# Load the data points
data = pd.read_csv('results/ls_data.csv')

# Load the found parameters
params = pd.read_csv('results/ls_params.csv')
a_found = params['a'][0]
b_found = params['b'][0]

plt.scatter(data['x'], data['y'], s=1, alpha=0.5, label='Data (with noise)')

x_vals = np.array([0, 1])
y_vals = a_found * x_vals + b_found
plt.plot(x_vals, y_vals, color='red', linewidth=2, label=f'Fit: y = {a_found:.2f}x + {b_found:.2f}')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Regression Fit')

plt.legend()
plt.savefig('results/ls_fit.png', bbox_inches='tight')
```

```python
# 1. Read the data
df = pd.read_csv('results/least_squares.txt')
# 2. Create the plot
plt.figure(figsize=(10, 6))
# 3. Get unique values of N and sort them
n_values = sorted(df['N'].unique())
for n in n_values:

    subset = df[df['N'] == n]
    
    # Plot threads vs time
    plt.plot(subset['threads'], subset['time_seconds'], marker='o', label=f'N = {n:,}')

plt.xlabel('Number of Threads')
plt.ylabel('Time (seconds) - Log Scale')
plt.title('Execution Time vs Threads for Least Squares')
plt.grid(True, which="both", ls="--", alpha=0.5)
plt.legend(title='Data Size (N)', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.yscale('log')

plt.xticks(sorted(df['threads'].unique()))
plt.tight_layout()

plt.savefig('results/least_squares.png', bbox_inches='tight')
```
